# 1. Make Some Detections

```python
#Membuka koneksi dengan kamera webcam. Argument 0 menunjukkan bahwa kita menggunakan kamera utama yang terhubung dengan komputer.
cap = cv2.VideoCapture(0)
#Menginisialisasi objek Holistic dari modul mp_holistic dengan parameter min_detection_confidence dan min_tracking_confidence yang menentukan kepercayaan minimum untuk deteksi dan pelacakan.
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    # Memulai loop untuk membaca setiap frame dari kamera selama kamera terbuka.
    while cap.isOpened():
        #Membaca frame terbaru dari kamera dan menyimpannya dalam variabel frame.
        ret, frame = cap.read()
        
        # Mengonversi warna frame dari BGR (yang digunakan oleh OpenCV) ke RGB (yang digunakan oleh Mediapipe).
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False       
        
        # hasil
        #Memproses frame menggunakan model holistic untuk mendapatkan hasil estimasi pose tubuh, deteksi wajah, dan deteksi tangan.
        results = holistic.process(image)
        # print(results.face_landmarks)
        
        # RGB - BGR
        image.flags.writeable = True   
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 1. Draw face landmarks
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                 mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                                 )
        
        # 2. Right hand
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                                 )

        # 3. Left Hand
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                                 )

        # 4. Pose Detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )
        #Menampilkan frame hasil pemrosesan di jendela yang bernama "Raw Webcam Feed".                
        cv2.imshow('Raw Webcam Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
# Inisialisasi holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    start_time = time.time()  # Waktu mulai
    
    while cap.isOpened():
        ret, frame = cap.read()
        
        # BGR - RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False        
        
        # Proses holistic
        results = holistic.process(image)
        
        # RGB - BGR untuk rendering
        image.flags.writeable = True   
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Gambar landmark wajah
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                 mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                                 )
        
        # Gambar landmark tangan kanan
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                                 )

        # Gambar landmark tangan kiri
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                                 )

        # Gambar landmark pose
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                                 mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )
       
        # Ekspor koordinat landmark ke CSV
        try:
            pose = results.pose_landmarks.landmark
            pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())
            
            face = results.face_landmarks.landmark
            face_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in face]).flatten())
            
            row = pose_row+face_row
            
            # Tambah class name 
            row.insert(0, class_name)
            
            # Ekspor ke CSV
            with open('dataset_isyarat.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(row) 
            
        except:
            pass
                        
        cv2.imshow('Raw Webcam Feed', image)

        # Cek apakah sudah 5 menit
        if time.time() - start_time >= 300:  # 300 detik = 5 menit
            break

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

## 3.1 Read in Collected Data and Process

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset_isyarat.csv')
```

```python
X = df.drop('class', axis=1) # features akan menjadi semua nilai koord
y = df['class'] # target = label kelas
```

```python
# X adalah fitur, y adalah target
# Misalkan X dan y sudah diinisialisasi sebelumnya

# Pisahkan data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=1234)
```

```python
# X_test adalah variabel yang berisi data testing setelah pemisahan
X_test
```

# 3.2 Train Machine Learning Classification Model

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 

from sklearn.ensemble import RandomForestClassifier
```

```python
pipelines = {

    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
}
#Ini akan melatih model Random Forest Classifier setelah data telah dinormalisasi menggunakan Standard Scaler.
```

```python
# Dictionary untuk menyimpan model-model yang sudah dilatih
fit_models = {}
# Iterasi melalui dictionary pipelines
for algo, pipeline in pipelines.items():
    # Training model dengan menggunakan pipeline pada data training
    model = pipeline.fit(X_train, y_train)
    # Menyimpan model yang sudah dilatih ke dalam dictionary fit_models
    fit_models[algo] = model
```

```python
# Dictionary untuk menyimpan model-model yang sudah dilatih seperti fungsi print
fit_models
```

```python
# Melakukan prediksi menggunakan model Random Forest yang sudah dilatih pada data testing
fit_models['rf'].predict(X_test)
```

## 3.3 Evaluate and Serialize Model 

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix # Accuracy metrics 
import pickle 
```

```python
# Iterasi melalui model-model yang sudah dilatih
for algo, model in fit_models.items():
    # Melakukan prediksi pada data testing menggunakan model yang sesuai
    y_prediksi = model.predict(X_test) #sebagai y prediksi
    # Menampilkan laporan klasifikasi untuk evaluasi performa model
    print(algo, classification_report(y_test, y_prediksi))
```
